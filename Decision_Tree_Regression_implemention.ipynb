```python
#Importing Libraries

import numpy as np 
import pandas as pd 
import matplotlib .pyplot as plt 
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
print("HII")
```

```python
#Importing dataset 
ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
#splliting dataset into Independent and dependent data 
x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
#fitting Decision Tree Regression model 
from sklearn.tree import DecisionTreeRegressor,plot_tree 
reg=DecisionTreeRegressor(random_state=0)
reg.fit(x,y)

```

```python

#Predict for  input 
myinput=pd.DataFrame(data=[[4]],columns=["Level"])
reg.predict(myinput)
```

```python
#Visualization  Decision Tree 
plt.figure(figsize=(50,40))
plot_tree(reg,feature_names=["Level"],filled=True)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
#Predict for user input 
level=float(input("Enter Level:"))
myinput=pd.DataFrame(data=[[level]],columns=["Level"])
result=reg.predict(myinput)
print(f"Predicted result for level {level}:{round(result[0])}")
```
